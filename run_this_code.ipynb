```python
#3.4
# part a
df = pd.read_csv('data/SGD_0.1.csv')
```

```python
epochs = df['Unnamed: 0']
training_loss = df['training_loss']
val_loss = df['val_losses']
train_accuracies = df['train_accuracies']
val_accuracies = df['val_accuracies']

plt.figure(figsize=(10, 5))
plt.plot(epochs, training_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss vs. Epochs')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(epochs, train_accuracies, label='Training Accuracy')
plt.plot(epochs, val_accuracies, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy vs. Epochs')
plt.grid(True)
plt.show()
```

```python
# part c
path_to_SGD = [
    'data/SGD_0.1.csv',
    'data/SGD_0.2.csv',
    'data/SGD_0.5.csv',
    'data/SGD_0.05.csv',
    'data/SGD_1.0.csv'
]
learning_rates = [0.1, 0.2, 0.5, 0.05, 1.0]

lowest_val_losses = {}

for path, lr in zip(path_to_SGD, learning_rates):
    df = pd.read_csv(path)
    lowest_val_loss = df['val_losses'].min()
    lowest_val_losses[lr] = lowest_val_loss

lowest_val_losses_df = pd.DataFrame(list(lowest_val_losses.items()), columns=['Learning Rate', 'Lowest Validation Loss'])

lowest_val_losses_df = lowest_val_losses_df.sort_values(by='Learning Rate')

plt.figure(figsize=(10, 5))
plt.plot(lowest_val_losses_df['Learning Rate'], lowest_val_losses_df['Lowest Validation Loss'], linestyle='-')
plt.xlabel('Learning Rate')
plt.ylabel('Lowest Validation Loss')
plt.title('Lowest Validation Loss for Different Learning Rates')
plt.grid(True)
plt.show()
```
